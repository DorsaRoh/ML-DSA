# Arrays

**Data structures store data in RAM (memory). RAM is where all variables are stored.**

- Bit = a position that can store a digit (either 0 or 1, i.e. binary)

- Byte = 8 bits

*Bit -> Bits -> Byte -> RAM -> Data structures*

There are two components to RAM: Value and Address.


```python
from IPython.display import Image, display
display(Image(filename='../assets/Arrays/1-ram.png', width=500, height=500))
```

Each integer occupies 4 bytes of space (4 * 8 bits), hence the addresses are 4 bytes apart.

```python
from IPython.display import Image, display
display(Image(filename='../assets/Arrays/2-ramchar.png', width=500, height=500))
```

Each character occupies 1 byte of space, hence the addresses are 1 byte apart.
